```python
import numpy as np
import matplotlib.pyplot as plt

# Function to find the epoch with the best score
def find_best_epoch(data):
    best_epoch = np.argmax(data) + 1  # Adding 1 because epochs start from 1
    best_score = np.max(data)
    return best_epoch, best_score

# Load the data from .npy files

bertscore_data = np.load(r'C:\Users\ayush\OneDrive\Documents\GitHub\Multi-document-summarization-for-medical-literature\Experiment\PICO-0.15\Generating_summary result\BERTscore.npy')
rouge1_data = np.load(r'C:\Users\ayush\OneDrive\Documents\GitHub\Multi-document-summarization-for-medical-literature\Experiment\PICO-0.15\Generating_summary result\rouge1.npy')
rouge2_data = np.load(r'C:\Users\ayush\OneDrive\Documents\GitHub\Multi-document-summarization-for-medical-literature\Experiment\PICO-0.15\Generating_summary result\rouge2.npy')
rougeL_data = np.load(r'C:\Users\ayush\OneDrive\Documents\GitHub\Multi-document-summarization-for-medical-literature\Experiment\PICO-0.15\Generating_summary result\rougeL.npy')

# Create subplots

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

# Function to plot data
def plot_data(ax, data, title):
    epochs = range(1, len(data) + 1)
    ax.plot(epochs, data)  # Epochs start from 1
    best_epoch, best_score = find_best_epoch(data)
    ax.scatter(best_epoch, best_score, color='red')
    ax.text(best_epoch, best_score, f'Best: {best_score:.2f} at epoch {best_epoch}')
    ax.set_title(title)
    ax.set_xlabel('Epochs')
    ax.set_ylabel('Score')
    ax.set_xticks(epochs)  # Explicitly set x-ticks
    ax.set_xticklabels(epochs)  # Explicitly set x-tick labels


#plot_data(axs[0, 0], bartscore_data, 'BERTscore')
# Plot BERTscore
plot_data(axs[0, 0], bertscore_data, 'BERTscore')

# Plot ROUGE-1
plot_data(axs[0, 1], rouge1_data, 'ROUGE-1')

# Plot ROUGE-2
plot_data(axs[1, 0], rouge2_data, 'ROUGE-2')

# Plot ROUGE-L
plot_data(axs[1, 1], rougeL_data, 'ROUGE-L')

# Show the plots
plt.tight_layout()
plt.show()
```
