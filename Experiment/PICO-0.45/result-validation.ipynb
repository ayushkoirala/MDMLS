```python
import numpy as np
import matplotlib.pyplot as plt

train_losses = np.load(r'C:\Users\ayush\OneDrive\Documents\GitHub\Multi-document-summarization-for-medical-literature\Experiment\PICO-0.45\train_losses.npy')


step_per_ep = train_losses.shape[-1]


flattened_train_losses = train_losses.ravel()


smoothed_losses = np.convolve(flattened_train_losses, np.ones(10) / 10, mode='same')

plt.figure(figsize=(10, 6))
plt.plot(range(step_per_ep * train_losses.shape[0]), smoothed_losses, label='Training loss')

for i in range(1, train_losses.shape[0]):
    plt.axvline(i * step_per_ep, linestyle='--', color='gray')


plt.xlabel('Training steps')
plt.ylabel('Loss (log scale)')
plt.title('Training losses')
plt.legend()

plt.yscale('log')


plt.show()
```
